# 相似度資料

```python
review_senti = pd.read_csv('review_senti_corr.csv')
review_senti
```

# 列出user和movie

```python
# all user
all_user = df['Author'].tolist()
all_user = list(set(all_user))
all_user
```

```python
# all movie
all_movie = df['Mid']
all_movie = list(set(all_movie))
all_movie
```

# 找出user沒評分過的movie

```python
# user評分過的電影
rated = df[df['Author'].str.contains(all_user[0])]['Mid'].tolist()
rated
```

```python
# 找出沒評論過的電影
test = list(set(all_movie)^set(rated))
test
```

```python
# 找出特定mid的行(找出其他電影的相似度)
find_m = review_senti[review_senti['Unnamed: 0'].str.contains(test[0])]
find_m
```

```python
# 找出相似度
for i in rated:
    print(find_m[i])
```
